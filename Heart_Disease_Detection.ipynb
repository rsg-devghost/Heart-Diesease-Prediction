Import the necessary dependencies


```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

Data collection and Processing

```python
#load the csv data
heart_data = pd.read_csv('/content/heart_disease_data.csv')
```

```python
# print first 5 rows of the dataset
heart_data.head()
```

```python
# print last 5 rows in ds
heart_data.tail()
```

The total number of rows and columns in the dataset.

```python
heart_data.shape
```

Getting some overview of the data

```python
heart_data.info()
```

Checking for missing values.
isnull() --> returns a dataframe object with true for NULL and false for else.
sum() --> returns a value of total sum of the parameter passed.

```python
heart_data.isnull().sum()
```

Statistical measure/insights of the data.

```python
heart_data.describe()
```

Checking the column[target] for count of 1's and 0's.
1 ---> Defective/Yes has problems.
0 ---> No/False problem.  

```python
heart_data['target'].value_counts()
```

Now we can extract/remove/seperate the target column from others.

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

Splitting data in train and test.

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.2, stratify = Y, random_state = 2 )
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

Model training

```python
model = LogisticRegression()
```

Train the mdoel with training data.

```python
model.fit(x_train.values,y_train.values)
```

Model evaluation

```python
x_train_prediction = model.predict(x_train)
train_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print('Accuracy in training data : ',train_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print('Accuracy in training data : ',test_data_accuracy)
```

FINALLY!!We can use the model to predict for input data!!

```python
input_data = (61,1,0,140,207,0,0,138,1,1.9,2,1,3)

# change the input data[csv] in numpy array, for ease of calculations
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we predict only for one instance, else it'll
# predict for 242 instances!!

input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)

print(prediction)

if (prediction[0] == 0):
  print('Person HAS heart diesease!!')

else:
  print('Person DOES NOT have heart diesease!!')
```
